1. Create a simple line plot for the following data:
  
  x = [1, 2, 3, 4, 5]
  y = [10, 15, 25, 30, 50]
  
  a. Plot the data.  
  b. Customize the plot by adding a title, axis labels, and a grid.  


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
```

```python
plt.plot(x,y,c='red')
plt.title("Trend of input and output data\n",fontdict={'fontsize':20,'fontweight':5,'color':'Maroon'})
plt.xlabel("Input",fontdict={'fontsize':12,'fontweight':5,'color':'red'})
plt.ylabel("Output",fontdict={'fontsize':12,'fontweight':5,'color':'red'})
ticks=np.arange(1,6,1)
plt.xticks(ticks)
plt.grid()
plt.show()
```

### Description

first i imported the libraries .To Create a simple line plot i assigned the given data x and y. I used plt.plot(x,y,c='red') to draw the line in red colour.then i added title and the used font dict to make the color ,size and weight of the font.and assigned the labels for x axis and y axis by using plt.xlabels and plt.ylabels again used the font dict to customize its appearence.I customized the x-axis tick marks using np.arange() and plt.xticks() setting ticks from 1 to 5.I included a grid in the background using plt.grid(), which makes the chart easier to read.finally I used plt.show() to view the plot.



2. Create a bar graph to represent the marks scored by students in a subject:  
  
  students = ['John', 'Jane', 'Alice', 'Bob']
  marks = [75, 85, 60, 90]
  
  a. Plot the data as a bar graph.  
  b. Customize the colors and add a title. 


```python
students = np.array(['John','Jane','Alice','Bob'])
marks = np.array([75,85,60,90])
```

```python
plt.bar(students,marks,width=0.7,align='center',color='purple')
plt.title("Students and their marks\n",fontdict={'fontsize':20,'fontweight':10,'color':'red'})
plt.xlabel("students",fontdict={'fontsize':12,'fontweight':5,'color':'red'})
plt.ylabel("Marks",fontdict={'fontsize':12,'fontweight':5,'color':'red'})
ticks=np.arange(0,110,10)
labels=["{}".format(i)for i in ticks]
plt.yticks(ticks,labels)
plt.show()
```

### Description

To Create a bar graph i assigned the given data students and marks. I used plt.bar(students,marks,width=0.7,align='center',color='purple')
 to draw the bar in purple color with the width of 0.7 .then i added title and the used font dict to customize the appearance of the font.and assigned the labels for x axis and y axis by using plt.xlabels and plt.ylabels again used the font dict .I customized the y-axis tick marks using np.arange() and plt.yticks() setting ticks from 0 to 100.finally I used plt.show() to view the plot.


3. Create a pie chart to represent the percentage distribution of a companys revenue from different regions:  
  
  regions = ['North America', 'Europe', 'Asia', 'Others']
  revenue = [45, 25, 20, 10]
  
  a. Create a pie chart with the region names as labels.  
  b. Highlight the region with the highest revenue.  


```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = [0.1 if val == max(revenue) else 0 for val in revenue]
```

```python
plt.pie(revenue,labels=regions,autopct="%.1f",explode = explode)
plt.title("percentage distribution of revenue in different regions\n",fontdict={'fontsize':20,'fontweight':5,'color':'red'})
plt.show()
```

```python
plt.pie(revenue,labels=regions,autopct="%.1f",explode = explode)
plt.title("percentage distribution of revenue in different regions\n",fontdict={'fontsize':20,'fontweight':5,'color':'red'})
plt.show()
```

### Description

To create a piechart I assigned the given data region and revenue.Iused explode to highlight the maximum percentage of revenue.i used plt.pie(revenue,labels=regions,autopct="%.1f",explode = explode) to view the piechart with the labels of region autopct for percentage and explode .I added title() and font dict to customize font.finally plt.show to view the result.


4. Generate a histogram to show the frequency distribution of randomly generated integers between 1 and 100 (sample size = 1000).

```python
data = np.random.randint(1, 101, size=1000) 
plt.hist(data,color='mediumvioletred',edgecolor='black')
plt.title("frequency distribution \n",fontdict={'fontsize':20,'fontweight':5,'color':'red'})
plt.show()
```

### Description

To create a histogram I assigned the given data using np.random.randint(1, 101, size=1000) to store random  1 to 100 numbers 
.I used plt.hist(data,color='mediumvioletred',edgecolor='black') to view the histogram in mediumvioletred color with the edge color black.I added title() and font dict to customize font.finally plt.show to view the result.
